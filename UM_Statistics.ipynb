**Population** **Variance**

```python
Population = [5,10,15,220,25,300,35,40,45]
```

```python
Population
```

```python
###pop_mean
pop_mean = sum(Population)/len(Population)
```

```python
pop_mean
```

```python
sq_pop_deviation = [(i-pop_mean)**2 for i in Population]
```

```python
sq_pop_deviation
```

```python
pop_variance = sum(sq_pop_deviation)/len(Population)
```

```python
pop_variance
```

```python
print('Population variance : {}'.format(round(pop_variance,2)))
```

Sample Variance

```python
import numpy as np
sample = np.arange(5,25,5)
```

```python
sample
```

```python
s_mean = sum(sample)/len(sample)
```

```python
s_mean
```

```python
sq_s_deviation = [(i-s_mean)**2 for i in sample]
```

```python
sq_s_deviation
```

```python
s_variance = sum(sq_s_deviation)/(len(sample)-1)
```

```python
s_variance
```

```python
print ('sample_variance: {} '.format(round(s_variance,2)))
```
